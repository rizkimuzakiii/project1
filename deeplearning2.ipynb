```python
from google.colab import files
uploaded = files.upload()  # Mengunggah file PDF
```

```python
import PyPDF2

def extract_text_from_pdf(pdf_file):
    # Use PdfReader instead of PdfFileReader
    reader = PyPDF2.PdfReader(pdf_file)
    text = ""
    for page_num in range(len(reader.pages)): #reader.pages replaces reader.numPages
        page = reader.pages[page_num] # Access pages using index, like a list
        text += page.extract_text()
    return text

# Ambil nama file yang diunggah
pdf_file_name = list(uploaded.keys())[0]

# Buka file menggunakan nama file yang diunggah
with open(pdf_file_name, 'rb') as f:
    document_text = extract_text_from_pdf(f)

print(document_text)  # Print hasil teks yang diekstrak
```

```python
from transformers import pipeline

# Load pre-trained model dari HuggingFace untuk Question Answering
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

# Definisikan fungsi untuk menjawab pertanyaan
def answer_question(question, context):
    result = qa_pipeline({
        'question': question,
        'context': context
    })
    return result['answer']

# Contoh penggunaan
question = "Apa yang dimaksud dengan Chatbot?"
context = document_text  # Menggunakan teks dari dokumen yang diunggah

answer = answer_question(question, context)
print(f"Answer: {answer}")
```

```python
def chat_bot():
    print("Chat Bot is ready! Type 'exit' to stop.")
    while True:
        question = input("Ask a question: ")
        if question.lower() == 'exit':
            break
        answer = answer_question(question, document_text)
        print(f"Answer: {answer}")

# Menjalankan chatbot
chat_bot()
```
